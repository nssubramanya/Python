### Select Data by Labels using .loc
**Note** .loc selects data only by index or column labels. Some data frames may not have meaningful names and default to integers from 0 to n-1, where n is length of data frame

#### Selecting a single row with .loc with a string

````python
mtcars.loc['Mazda RX4']

# This returns a single row of data with all columns and type. 
# This returns the row of data as a series
````

#### Selecting multiple rows with .loc using list of strings
This returns a Data frame with rows in the order specified in the list

````python
mtcars.loc[['Mazda RX4', 'Toyota Corolla', 'Toyota Corona']]
````

#### Selecting multiple rows with .loc using Slice Notation [:]
Slice Notation is defined by [start:stop:step] values. Note that pandas includes the stop value unlike python lists. 

````python
mtcars.loc['Mazda RX4':'Datsun 710']
````

````python
mtcars.loc['Mazda RX4':'Toyota Corolla':3]
````

#### Select Columns using .loc

````python
mtcars.loc['Mazda RX4',['Mpg', 'Horse Power']]
````

## Selecting data by integer location using .iloc
* Every row and column of data in a DataFrame has an integer location that defines it
* The integer location is the number of rows/columns from the top/left beginning at 0
* We can use an integer, list of integers or slice notation with .iloc
* .iloc can be used to fetch rows or columns

#### Selecting a single row with .iloc with an integer
This returns the row as a Series

````python
mtcars.iloc[0]
````

#### Selecting multiple rows with .iloc with list of integers

````python
# We are selecting 1st and last rows
# This returns a DataFrame and not a series
mtcars.iloc[[0,-1]] 
````

#### Selecting multiple rows with .iloc using slice notation

````python
mtcars.iloc[0:5:2]
````

### Selecting both rows and columns with .iloc and .loc
Separate Row and column specifier with a comma (,)

````python
# Get the first 3 rows and 3 Columns
mtcars.iloc[0:3, 0:3]
````

````python
# Get Named Rows and Columns
mtcars.iloc[[1,5,10], 1:5]
````

## Mixing Label strings and Numbers in .loc/.iloc
* .ix was initailly used for mixing labels and integer locations. This was confusing and ambiguous, so has been depricated!
* If we need to mix Integers and Labels, convert both to a single type - either both Labels or both Integers

````python
col_names = mtcars.columns[[0, 1,5,7]]
col_names
mtcars.loc['Mazda RX4':'Datsun 710', col_names]
````

````python
labels = ['Mazda RX4', 'Mazda RX4 Wag']
mazda_indeces = [mtcars.index.get_loc(label) for label in labels]
mazda_indeces

mtcars.iloc[mazda_indeces, :]
````

````python
# all Merc vehicles
vehicle_indeces = [mtcars.index.get_loc(s) for s in mtcars.index if 'Merc' in s]
vehicle_indeces
mtcars.iloc[vehicle_indeces,:]
````
